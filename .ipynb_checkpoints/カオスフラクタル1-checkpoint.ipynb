## ロジスティック写像

$$x_{n+1} = a x_{n}(1-x_n)$$


```python
import numpy as np
import matplotlib.pyplot as plt

a = 2.0
x = [0.1]
x_1 = [] # 更新するxの値を保存しておく配列
for i in x:
  y = x[len(x)-1] # 
  z = a*y * (1-y) # 更新値
  x.append(z) # リターンマップで描画するx軸の値は、直前のxの値である。よってxの値も保存しておく
  x_1.append(z) # リターンマップのy軸に描画する配列。
  
  if len(x) > 25:
    break
print(x)
x_axis = np.arange(0, len(x), 1)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].set_title("return map, a=2.0")
axes[0].scatter(x[0:len(x)-1], x_1)
axes[0].set_xlabel("$x_n$")
axes[0].set_ylabel("$x_{n+1}$")

axes[1].set_title("time series a=2.0")
axes[1].set_ylabel("$x_n$")
axes[1].set_xlabel("n")
axes[1].plot(x_axis, x)
```
