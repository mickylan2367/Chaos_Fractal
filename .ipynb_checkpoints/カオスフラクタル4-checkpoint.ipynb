```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def absolute(x):
    y = []
    for p in x:
        if p<0:
            p = -p
        y.append(p)
    return y

def absolute(x):
    mask = x<0
    x[mask] = -x[mask]
    return x

x = np.array([1, 2, -3])
y = absolute(x)
y
```

```python
def logistic(a):
    # 関数を先に作っておくほうが、プログラム内で何をしているのか確かにわかりやすいかも...    
    x = [0.8]
    for i in range(1000):
        x.append(a * x[-1] * (1 - x[-1]))
    return x # 後ろから100番目以降の配列を取得（収束済みと踏んでいる）

def Lyapunov_exponent(a, x):
    y = np.array([np.log(np.abs(a*(1-2*i))) for i in x])
    sum_y = np.nansum(y)
    sum_y /= len(x)
    return sum_y

fig, axes = plt.subplots(1, 2, figsize=(16, 8))
li = []
for a in np.linspace(2.8, 4.0, 4000):
    x = logistic(a)
#     print(x)
    li.append(Lyapunov_exponent(a, x))
    axes[0].plot([a]*len(x[-50:]), x[-50:], "c.", markersize=1.7) 
    # なるへそ、その場でプロットしているのね
    # でもこの状態でプロットちゃうと、収束点には大量の点が一か所にプロットされる感じになるけど、...ま、それでもいいのかな

li_x = np.linspace(2.8, 4.0, 4000)
axes[1].plot(li_x, li, "m.", markersize =1.7)

axes[0].set_xlabel("a")
axes[0].set_ylabel("x_n")
axes[1].set_xlabel("a")
axes[1].set_ylabel("Lyapunov_exponent")
plt.show()
```

* カオスかどうかはリアプノフ指数を見て決める
* 妙にリアプノフ指数が落ちている場所は、周期解が発生している.一回目にa=3.2..???ぐらいで発生しているのは2周期解における収束
