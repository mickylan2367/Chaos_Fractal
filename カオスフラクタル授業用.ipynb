```python
import numpy as np
import matplotlib.pyplot as plt

a = 2.0
x = [0.1]
x_1 = []
for i in x:
  y = x[len(x)-1]
  z = a*y * (1-y)
  x.append(z)
  x_1.append(z)
  
  if len(x) > 25:
    break
print(x)
x_axis = np.arange(0, len(x), 1)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].set_title("return map, a=2.0")
axes[0].plot(x[0:len(x)-1], x_1)
axes[1].set_title("time series a=2.0")
axes[1].plot(x_axis, x)
```
